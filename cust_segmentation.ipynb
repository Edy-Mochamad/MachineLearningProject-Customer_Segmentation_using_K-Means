### **K-Means Clustering**

```python
data.columns
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
kmeans.fit(X)
```

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit_predict(X)
```
